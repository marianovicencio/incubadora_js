```python
# read csv called "Potencia20_punto4" and plot the columns Temperatura vs Date
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Potencia20_punto4.csv", delimiter=';')
temperatura = df['Temperatura']
tiempo = df['Date']
plt.plot(tiempo, temperatura)   
plt.xlabel('Date')
plt.ylabel('Temperatura')
plt.title('Temperatura vs Date')
plt.show()

# suavizar la curva
import numpy as np
from scipy.signal import savgol_filter
temperatura = savgol_filter(temperatura, 51, 3)
plt.plot(tiempo, temperatura)
plt.xlabel('Date')
plt.ylabel('Temperatura')
plt.title('Temperatura vs Date')
plt.show()

```
