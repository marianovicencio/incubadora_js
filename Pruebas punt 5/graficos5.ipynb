```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("100V100A.csv", delimiter=';')
temperatura = df['Temperatura']
tiempo = df['Date']
df['Date'] = pd.to_datetime(df['Date'], format='%H:%M:%S')
plt.scatter(df['Date'], temperatura)
# Formatear el eje X para que muestre menos etiquetas (cada 5 minutos, por ejemplo)
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=5))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.xticks(rotation=45)
plt.xlabel('Tiempo')
plt.ylabel('Temperatura')
plt.title('Temperatura vs Tiempo')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("50V20A.csv", delimiter=';')
temperatura = df['Temperatura']
tiempo = df['Date']
plt.scatter(tiempo, temperatura)
num_ticks = 9  
tick_positions = tiempo[::len(tiempo)//num_ticks]  
tick_labels = tick_positions  
plt.xticks(tick_positions, tick_labels, rotation=45)
plt.xlabel('Tiempo')
plt.ylabel('Temperatura')
plt.title('Temperatura vs Tiempo')
plt.tight_layout()
plt.show()
```
